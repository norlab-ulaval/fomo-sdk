# Audio Data Example

This notebook demonstrates how to load, visualize, and manipulate Audio data using the `fomo_sdk`.

```python
import matplotlib.pyplot as plt
from fomo_sdk.audio.utils import Stereo
```

## 1. Load Audio Data

We will load audio data from the example dataset.

```python
dataset_base_path = "data"
deployment = "2025-10-14"
trajectory = "red"

stereo_data = Stereo()
stereo_data.load_fomo_audio(dataset_base_path, deployment, trajectory)
```

## 2. Visualization

We can create a spectrogram of the audio data.

```python
image = stereo_data.create_spectogram(show=True)
```

We can also plot the raw audio channels.

```python
idx_start = 0
idx_end = 1000

plt.plot(
    stereo_data.left_timestamps[idx_start:idx_end],
    stereo_data.left[idx_start:idx_end],
    label="left",
)
plt.plot(
    stereo_data.right_timestamps[idx_start:idx_end],
    stereo_data.right[idx_start:idx_end],
    label="right",
)
plt.legend()
plt.show()
```

## 3. Saving audio data
The audio data in FoMo dataset are split into 1 second-long segments split between left and right channel. Users can choose to export the data as a single double channel audio file.

```python
stereo_data.save_audio("data/audio.wav", True)
```

## 4. Playing audio data
We can either play an individual segment or the entire audio file.

```python
from IPython.display import Audio, display

# Play a 1-second segment of the left channel
print("Playing 1-second segment of left channel:")
display(
    Audio(
        data=stereo_data.left[: stereo_data.sample_rate], rate=stereo_data.sample_rate
    )
)

# Play the complete exported audio file
print("Playing exported stereo file:")
display(Audio(filename="data/audio_stereo.wav"))
```
